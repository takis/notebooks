```python
%pylab inline
data = [0.3, 0.2, 0.4, 0.1]
p = pie(
    data,
    explode=[0,0.1,0,0],
    labels=['konijn', 'kip', 'koe', 'kameel'],
    colors=('#ffaa00', '#771234', '#fa12bc'),
    shadow=True,
    labeldistance=0.5,
    startangle=90,
    radius=1.1,
    counterclock=False,
    wedgeprops={
        'linewidth': 3,
        'linestyle': 'dashed',
        'alpha': 0.9,
        'capstyle': 'round',
        'edgecolor': 'white',
        'joinstyle': 'bevel',
        'fill': True,
        'hatch': 'o',
    },
    textprops={
        'alpha': 0.9,
        'linespacing': 2.0,
        'family': 'monospace',
        'rotation': 0,
        'fontsize': 20,
        'style': 'italic',
        'weight': 'light',
        'backgroundcolor': 'red',
    },
)
```
